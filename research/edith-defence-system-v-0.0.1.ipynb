```python
from ultralytics import YOLO
import torch
import shutil

def train_model():
    # Check if CUDA is available
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Using device: {device}")

    # Load a YOLOv8 model
    model = YOLO('yolov8n.pt')  # load pretrained model (recommended for training)

    # Train the model with our custom dataset
    results = model.train(
        data='/kaggle/input/data-yaml/data.yaml',  # Path to data.yaml file
        epochs=105,                 # Number of epochs
        imgsz=640,                  # Image size
        batch=16,                   # Batch size
        device=device,              # CUDA device (GPU)
        workers=4,                  # Number of worker threads
        patience=20,                # Early stopping patience
        save=True,                  # Save checkpoints
        project='runs/train',       # Project name
        name='yolov8n_birds_drones'  # Experiment name
    )

    # Validate the model
    results = model.val()

    # Compress the runs folder into a ZIP file
    zip_folder('runs', 'runs.zip')
    print("Training complete. Runs folder compressed to runs.zip for download.")

def zip_folder(source_folder, output_zip):
    """Compresses a folder into a zip file."""
    shutil.make_archive(output_zip.replace('.zip', ''), 'zip', source_folder)
    print(f"Folder {source_folder} has been zipped as {output_zip}")

if __name__ == "__main__":
    train_model()
```
